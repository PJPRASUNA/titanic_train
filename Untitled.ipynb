# Exploratory Data Analysis




## IMPORTING LIBRARIES

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

## Reading the Dataset titanic_train

```python
file=pd.read_csv('titanic_train.csv')
file.head(10)
```

# Finding Null values

```python
file.isnull()
```

# Counting Null values 

```python
count_of_missing_values=file.isnull().sum()
count_of_missing_values[0:12]
```

# Finding datatype

```python
file.Age.dtype
```

```python
file.Fare.dtype
```

# Filling Nan values with other values 

```python
file.Cabin.fillna('No Cabin found')
```

# Replacing 

```python
file.Embarked.replace('S','P')
```

# Line Plot Given Data

```python
plt.figure(figsize=(14,6))
plt.title('Line Plot for Age and Fare')
sns.lineplot(data=file,x='Age',y='Fare')
plt.show()
```

# Bar Plot for given data 

```python
plt.figure(figsize=(14,6))
plt.title('Bar Plot for Survived and Age')
sns.barplot(data=file,y='Survived',x='Age')
```

# Heat Map for given data 

```python
plt.figure(figsize=(14,6))
plt.title('Heat Map for given data')
sns.heatmap(file.isnull())
```

# Distplot for given data 

```python
plt.figure(figsize=(14,6))
sns.distplot(file['Survived'].dropna(),color='blue',bins=10)
plt.title('Displot for given data')
```

# Box plot for given data 

```python
plt.figure(figsize=(14,6))
plt.title('Box Plot for given data')
sns.boxplot(x='Embarked',y='Age',data=file,palette='Dark2_r')
```

# Scatter Plot for given data

```python
plt.figure(figsize=(14,6))
plt.title('Scatter Plot for given data')
sns.scatterplot(x='Pclass',y='Age',data=file,palette='Dark2_r')
```

# Pair Plot for given data 

```python
plt.figure(figsize=(14,6))
sns.pairplot(file)
```

# Cleaning dataset

```python
file.drop('Age',axis=1,inplace=True)
```
